```python
# Initialize Faker
faker = Faker('en_US')
Faker.seed(4322)
lifecycle_stage = ['Subscriber', 'Lead', 'Marketing Qualified Lead', 'Sales Qualified Lead', 'Opportunity', 'Customer', 'Evangelist', 'Other']
lead_status = ['New', 'Open', 'In Progress', 'Open Deal', 'Unqualified', 'Attempted to Contact', 'Connected', ' Bad Timing']
num_records=1000
```

```python
companies_df=pd.read_csv('hubspot-crm-exports-all-companies-2025-10-23-2.csv')
```

```python
def generate_contact_data(num_records):
  contacts = []
  """Generate sample company data using Faker library"""
  for i in range(num_records):
    random_company = companies_df.sample(n=1)
    record_id = random_company['Record ID'].iloc[0]
    company_domain = random_company['Company Domain Name'].iloc[0]
    contact_firstname = faker.first_name()
    contact_lastname = faker.last_name()
    contact_email = contact_firstname.lower()+'.'+contact_lastname.lower()+'@'+company_domain
    contact_job = faker.job()
    contact_phone = faker.numerify(text='+1(%##)###-####')
    lifecycle_stages = random.choice(lifecycle_stage)
    lead_statuses = random.choice(lead_status)
    contact_country = faker.current_country()
    contact_state = faker.state()
    street_address = faker.street_address()
    city = faker.city()
    postal_code = faker.postcode()

    contact = {
              "Company Record ID": record_id,
              "Email": contact_email,
              "First Name": contact_firstname,
              "Last Name": contact_lastname,
              "Contact Owner": 'ABHINANDAN MISHRA',
              "Job Title": contact_job,
              "Phone Number": contact_phone,
              "Lifecycle Stage": lifecycle_stages,
              "Lead Status": lead_statuses,
              "Country/Region": contact_country,
              "State/Region": contact_state,
              "Street Address": street_address,
              "City": city,
              "Postal Code": postal_code

          }

    contacts.append(contact)
  return contacts
```

```python
def export_to_csv(contacts, filename):
  """Export company data to CSV file using pandas"""
  df = pd.DataFrame(contacts)
  timestamp_str = str(current_timestamp).replace('-', '').replace(':', '').replace(' ', '').replace('.', '')
  filename = f"{filename}_{timestamp_str}.csv"
  df.to_csv(filename, index=False)
  print(f"\nData exported to {filename}")
  return df
```

```python
current_timestamp = datetime.now()
hs_contacts=generate_contact_data(num_records)
export_to_csv(hs_contacts, "hs_contacts")
```

```python
contacts_df=pd.read_csv('hs_contacts_20251025125331867082.csv')
# Add 'con_' prefix to all column names in contacts
contacts_df.columns = ['con_' + col for col in contacts_df.columns]
companies_df.columns= ['acc_' + col for col in companies_df.columns]
# Display the first few rows with the new column names to verify
display(contacts_df.head())
display(companies_df.head())
```

```python
# Merge the two dataframes on the specified columns
merged_df = pd.merge(companies_df, contacts_df, left_on='acc_Record ID', right_on='con_Company Record ID', how='inner')

# Display the first few rows of the merged dataframe to verify
display(merged_df.shape[0])
```

```python
# Get the count of null values for each column in merged_df
null_counts = merged_df.isnull().sum()

# Filter for columns with at least one null value
columns_with_nulls = null_counts[null_counts == 1000]

# Print the columns and their null counts
print("Columns with null values and their counts:")
display(columns_with_nulls)
```

```python
# Get the list of columns to drop (those with 1000 null values)
columns_to_drop = columns_with_nulls.index.tolist()

# Drop the columns from the merged_df
merged_df = merged_df.drop(columns=columns_to_drop)

# Display the first few rows of the updated dataframe to verify
display(merged_df.head())
```

```python
# Filter for rows where 'con_Company Record ID' is null
null_con_records_df = merged_df[merged_df['con_Company Record ID'].isnull()]

# Count the number of such records
null_con_records_count = null_con_records_df.shape[0]

# Print the count
print(f"The number of records where 'con_Company Record ID' is null is: {null_con_records_count}")
```
